```python
#KDEplot
plt.figure(figsize=(8,5))
sns.kdeplot(data['Sales'], color = 'red', label = 'Sales', shade = True, bw=20)
sns.kdeplot(data['Profit'], color = 'blue', label = 'Profit', shade = True, bw=20)
plt.legend()               #legend() is an area describing the elements of the graph
```

Kernel Density Estimate(KDE) is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

**From the above graph we can conclude that, Profit is more than that of sales, but there are some areas where profit could be increased.**

**Segment wise  analysis of Profit, Discount and Sales**

```python
df = data.groupby(['Segment'])[['Profit','Discount','Sales']].mean()
df
```

```python
data.groupby(['Segment'])[['Profit','Discount','Sales']].mean().plot(kind = 'pie', subplots = True,startangle=90, figsize=(15,10),autopct='%1.1f%%')
plt.figure(figsize=(10,15))
plt.show()
```

**Sales:**

**Therefore,**

Home-office has High Sales and High Profit

**State wise  analysis of Profit, Discount and Sales**

```python
df = data.groupby(['State'])[['Profit','Discount','Sales']].mean()
df.head(10)
```

```python
state_profit = df.sort_values('Profit')
state_profit[['Profit']].plot(kind = 'bar',figsize = (15,10))
plt.title('State-wise Profit analysis',fontsize = 20)
plt.xlabel('States', fontsize = 20)
plt.ylabel('Profit',fontsize = 20)
plt.show()
```

**Highest Profit** = Vermont

**Lowest Profit** = Ohio

```python
state_sales = df.sort_values('Sales')
state_sales[['Sales']].plot(kind = 'bar',figsize = (15,10))
plt.title('State-wise Sales analysis',fontsize = 20)
plt.xlabel('States', fontsize = 20)
plt.ylabel('Sales',fontsize = 20)
plt.show()
```

**Highest amount of Sales :** Wyoming

**City wise analysis of Profit**

```python
df = data.groupby(['City'])[['Profit']].mean()
df.head(10)
```

```python
city_profit = df.sort_values('Profit').head(20)
city_profit[['Profit']].plot(kind = 'bar',figsize = (10,8))
plt.title('City-wise Profit analysis',fontsize = 20)
plt.xlabel('City', fontsize = 20)
plt.ylabel('Profit',fontsize = 20)
plt.show()
```

**Lowest Profit** = Bethlehem
